```python
from ntscraper import Nitter
import pandas as pd
```

```python
scraper = Nitter(0)
```

```python
terms = ["Hey Banco"]
```

```python
tweets = scraper.get_tweets(terms, mode='term')
tweets
```

```python
print(tweets.keys())
```

```python
keys = list(tweets.keys())
if len(keys) > 1:
    tweet = tweets[keys[1]]  # Access the second available key
else:
    print("Not enough keys in the dictionary.")
```

```python
tweets[1]
```

```python
recopilation_tweets = []
for tweet in tweets['tweets']:  # Directly access 'tweets' key in the tweets dictionary
    data = [tweet['user']['username'], tweet['text'], tweet['date']]
    recopilation_tweets.append(data)
```

```python
dataframe = pd.DataFrame(recopilation_tweets, columns = ['username', 'tweet', 'date'])
filtered_dataframe = dataframe[dataframe['username'] != '@heybanco']
filtered_dataframe
```
